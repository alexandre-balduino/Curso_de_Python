# Desafio 005

Você recebeu dois relatórios separados do sistema da empresa. O primeiro relatório contém as Vendas do mês, mas possui apenas o ID_Cliente. O segundo relatório é o Cadastro de Clientes, que liga o ID_Cliente ao Nome e à Região dele.

Seu objetivo é cruzar essas duas tabelas para gerar um relatório unificado e responder a algumas perguntas da diretoria.

### Seu código inicial:



```python
import pandas as pd

# Tabela A: Movimentação de Vendas
dados_vendas = {
    'ID_Cliente': [101, 102, 101, 103, 102, 104],
    'Produto': ['Notebook', 'Mouse', 'Teclado', 'Monitor', 'Notebook', 'Teclado'],
    'Preço': [4000.00, 150.00, 250.00, 1200.00, 4000.00, 250.00],
    'Quantidade': [1, 2, 3, 1, 1, 5]
}
df_vendas = pd.DataFrame(dados_vendas)

# Tabela B: Cadastro de Clientes
dados_cadastro = {
    'ID_Cliente': [101, 102, 103, 104],
    'Nome_Cliente': ['Alexandre', 'Vivian', 'Estevão', 'Ester'],
    'Região': ['Sul', 'Sul', 'Sudeste', 'Nordeste']
}
df_cadastro = pd.DataFrame(dados_cadastro)

print("Tabela de vendas:")
display(df_vendas)
print("Tabela de Cadastro:")
display(df_cadastro)
```

### Suas tarefas:

1. **O Cruzamento**: Junte os dois DataFrames em um único chamado df_final usando como base a coluna em comum.

```python
df_final = pd.merge(df_vendas, df_cadastro, on="ID_Cliente", how="inner")
display(df_final)
```

2. **Faturamento por Produto**: Utilizando o df_final, calcule e mostre na tela o Faturamento Total de cada produto. Adicione essa informação como uma nova coluna chamada 'Faturamento'.

```python
df_final["Faturamento"] = df_final["Preço"] * df_final["Quantidade"]
display(df_final)
```

3. **Análise Regional**: Descubra qual Região gerou o maior faturamento total para a empresa. Agrupe os dados por 'Região' e some o 'Faturamento'.

```python
display(df_final.groupby('Região')['Faturamento'].sum())
```

4. **Filtro de Elite**: Exiba uma tabela contendo apenas as vendas onde o faturamento por linha foi maior que R$ 1.000,00.

```python
display(df_final[df_final['Faturamento'] > 1000])
```

# Desafio 003

Sua missão neste desafio é pegar a nossa mesma tabela de informática e resolver 3 problemas práticos de alteração e resumo de dados.

---

### Seu código inicial:

```python
import pandas as pd

dados_informatica = {
    'Produto': ['Notebook Dell', 'Teclado Mecânico', 'Monitor LG 29"', 'Mouse Gamer', 'Impressora HP', 'SSD 1TB'],
    'Categoria': ['Notebooks', 'Periféricos', 'Monitores', 'Periféricos', 'Printers', 'Armazenamento'],
    'Preço': [4200.00, 250.00, 1350.00, 180.00, 890.00, 410.00],
    'Estoque': [5, 15, 4, 22, 3, 18]
}

df = pd.DataFrame(dados_informatica)
display(df)
```

### Suas tarefas:

1. **Correção de Dados**: O estagiário cadastrou a categoria da Impressora HP como 'Printers' (em inglês). Modifique esse valor específico para ficar em português: 'Periféricos'.

```python
print(f"A categoria da linha 4 era '{df.loc[4, 'Categoria']}' e foi modificada para ", end="")
df.loc[4, "Categoria"] = "Periféricos"
print(f"'{df.loc[4, "Categoria"]}'")
```

2. **A Inflação Chegou** (Operação em Massa): Devido às taxas de importação, todos os produtos da loja sofreram um aumento de 10% no preço. Atualize a coluna 'Preço' refletindo esse novo valor para todos os produtos de uma só vez.

```python
print("Os preços eram:")
display(df[['Produto', 'Preço']])
df["Preço"] = df["Preço"] * 1.1
print("Receberam 10% de aumento e foram para:")
display(df[["Produto", "Preço"]])
```

3. **Resumo por Categoria**: O gerente quer saber quantos produtos a loja tem no estoque total somando por cada Categoria. Agrupe os dados e mostre essa soma.

```python
display(df.groupby('Categoria')['Estoque'].sum())
```

### Suas Tarefas:

1. Use o módulo nativo `os` do Python para simular a leitura de variáveis de ambiente.

```python
import os
from dotenv import load_dotenv

load_dotenv()

db_host = os.environ.get("DB_HOST")
db_user = os.environ.get("DB_USER")
db_pass = os.environ.get("DB_PASS") or ""
db_data = os.environ.get("DB_NAME")
```

2. Crie uma função chamada `conectar_banco()` que simula o teste de conexão.

```python
from sqlalchemy import create_engine

def conectar_banco(
    host,
    user,
    password,
    database
):
    url_conexao = f"mysql+mysqlconnector://{user}:{password}@{host}/{database}"
    engine = create_engine(url_conexao)
    return engine
```

3. Leia os dados da nossa tabela de visitas simulando uma consulta remota

```python
import pandas as pd

os.system("mariadbd-safe &")

conexao = conectar_banco(
    db_host,
    db_user,
    db_pass,
    db_data
)

def ler_sql(cmd, cnx):
    try:
        resposta = pd.read_sql(
            cmd,
            cnx
        )
        return resposta
    except Exception as e:
        print(f"Erro: {e}")

atendimentos = ler_sql(
    "SELECT * FROM atendimentos",
    conexao
)
pacientes = ler_sql(
    "SELECT * FROM pacientes",
    conexao
)

tabela_inexistente = ler_sql(
    "SELECT * FROM tabela",
    conexao
)

display(atendimentos)
display(pacientes)
```

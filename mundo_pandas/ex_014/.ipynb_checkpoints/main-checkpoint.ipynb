### Suas Tarefas de Análise Visual:

1. **Fusão Confiável** (1:N): Cruze os DataFrames. Converta Data_Hora garantindo que o dia venha primeiro: pd.to_datetime(df['Data_Hora'], dayfirst=True).

```python
df_atendimentos["Data_Hora"] = pd.to_datetime(
    df_atendimentos["Data_Hora"],
    dayfirst=True
)

display(df_atendimentos.info())
```

```python
df_consolidado = pd.merge(
    df_pacientes,
    df_atendimentos,
    on="ID_Paciente",
    how="inner"
)

df_consolidado.head()
```

2. **Gráfico 1**: Crie um gráfico de barras verticais do Custo Total Acumulado por Agravo. Personalize-o (adicione título, cor nas barras e rotacione os nomes no eixo X para não encavalarem). Salve o arquivo final como grafico_custo.png.

```python
import matplotlib.pyplot as plt
```

```python
custo_total = df_consolidado.groupby("Agravo")["Custo_Atendimento"].sum()

plt.bar(
    custo_total.index,
    custo_total.values,
    color="forestgreen"
)
plt.title("Custo Total por Agravo")
plt.show()
```

3. **Gráfico 2**: Agrupe os dados por mês (df['Data_Hora'].dt.to_period('M')), some os custos, e faça o matplotlib desenhar um gráfico de linha do custo mensal.

```python
df_consolidado["Mes"] = df_consolidado["Data_Hora"].dt.strftime("%B")
custo_mensal = df_consolidado.groupby("Mes")["Custo_Atendimento"].sum()

plt.bar(
    custo_mensal.index,
    custo_mensal.values,
    color="forestgreen"
)
plt.xticks(rotation=45)
plt.title("Custo Total por Mês")
plt.show()
```

### As 3 perguntas que o pandas deve responder:

* **Pergunta 1** (A Regra dos Crônicos):

  "Quais pacientes (Nome e Idade) retornaram ao hospital 3 ou mais vezes no acumulado do ano? Precisamos incluí-los no nosso programa de acompanhamento preventivo de hipertensos/crônicos."

  Dica Pandas: Conte as ocorrências de ID_Paciente na tabela de atendimentos, filtre os que têm contagem => 3 e faça o merge com a tabela de pacientes para obter o Nome e a Idade.

```python
cronicos = df_consolidado.groupby("ID_Paciente").filter(lambda x: len(x) >= 3).drop_duplicates(subset="ID_Paciente")
display(cronicos[["Nome", "Idade"]])
```

* **Pergunta 2** (O faturamento por faixa etária):

  "Qual faixa etária está gerando o maior custo de internação para o hospital: Jovens (Até 30 anos), Adultos (de 31 a 59 anos) ou Idosos (60 anos ou mais)?"

```python
df_consolidado["Faixa_Etaria"] = pd.cut(
    df_consolidado["Idade"],
    bins=[0, 12, 18, 30, 60, 120],
    labels=["Criança", "Adolescente", "Jovem Adulto", "Adulto", "Idoso"],
    right=False
)
custo_faixa_etaria = df_consolidado.groupby("Faixa_Etaria")["Custo_Atendimento"].sum()
print(f"A faixa etária com maior custo de internação foi {custo_faixa_etaria.index.max()} com um total de R${custo_faixa_etaria.values.max():.2f} no acumulado do período.")
```

* **Pergunta 3** (O Gargalo de Atendimento):

  "Qual é o mês do ano em que registramos o maior número absoluto de atendimentos (volume de visitas, não custo)? Precisamos dessa informação para planejar a escala de contratação de médicos temporários."

```python
mes_mais_atendimentos = df_consolidado.groupby("Mes")["ID_Atendimento"].count()
print(f"O mês do ano com maior número de atendimentos foi {mes_mais_atendimentos.index.max()} com {mes_mais_atendimentos.values.max()} atendimentos")
```

* **Pergunta 4** (Os Pacientes Multidiagnóstico):

  "A diretoria quer saber: existem pacientes que foram atendidos por mais de um tipo de agravo diferente ao longo do ano? Quem são eles (Nome e Idade) e quais foram os agravos que cada um teve?"

```python
multidiagnostico = df_consolidado.groupby("ID_Paciente").filter(lambda x: x["Agravo"].nunique() > 1)
multidiagnostico = multidiagnostico.groupby(["ID_Paciente", "Nome", "Idade"])["Agravo"].apply(lambda x: ", ".join(x.unique())).reset_index()
display(multidiagnostico[["Nome", "Agravo"]])
```

* **Pergunta 5** (Tratamento de Alto Custo na Terceira Idade):

  "Precisamos auditar as contas. Liste apenas os pacientes idosos (60 anos ou mais) que tiveram pelo menos um atendimento individual com custo acima da média geral de todos os atendimentos do hospital."

  * Onde o bicho pega: Primeiro você precisa calcular a média global de Custo_Atendimento. Depois, aplicar um filtro duplo (Idade >= 60 E Custo > Média Global). No final, certifique-se de que cada paciente apareça apenas uma vez na lista (.drop_duplicates()).

```python
idoso_alto_custo = df_consolidado[(df_consolidado["Idade"] >= 60) & (df_consolidado["Custo_Atendimento"] > df_consolidado["Custo_Atendimento"].mean())].drop_duplicates()
display(idoso_alto_custo[["Nome", "Idade", "Agravo", "Custo_Atendimento"]])
```

* **Pergunta 6** (A Regra da Reinfecção / Retorno pelo Mesmo Motivo):

  "Queremos identificar falhas no tratamento ou surtos recorrentes. Quais pacientes retornaram ao hospital pelo exato mesmo agravo? (Exemplo: Carlos foi por Dengue em Janeiro e voltou por Dengue em Junho)."

  * Onde o bicho pega: Você terá que agrupar os dados usando duas colunas ao mesmo tempo: ['ID_Paciente', 'Agravo']. Use o seu querido .filter(lambda x: ...) para manter apenas os grupos que possuem tamanho maior que 1 (ou seja, onde houve repetição do mesmo problema para o mesmo paciente).

```python
reincidentes = df_consolidado.groupby(["ID_Paciente", "Agravo"]).filter(lambda x: len(x) > 1)[["ID_Paciente", "Nome","Agravo"]]
reincidentes["Reincidência"] = reincidentes.groupby(["ID_Paciente", "Agravo"])["Agravo"].transform("count")
reincidentes = reincidentes.drop_duplicates(subset=["ID_Paciente", "Agravo"])
display(reincidentes)
```

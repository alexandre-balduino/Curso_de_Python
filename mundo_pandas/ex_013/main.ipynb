### Suas Tarefas Obrigatórias:

1. **Fusão e Tratamento de Tempo**: Faça o merge dos DataFrames e converta a coluna Data_Hora para o tipo datetime real usando pd.to_datetime(df['Data_Hora'], format='mixed').

```python
df_atendimentos["Data_Hora"] = pd.to_datetime(
    df_atendimentos["Data_Hora"],
    dayfirst=True
)

df_consolidado = pd.merge(
    df_pacientes,
    df_atendimentos,
    on="ID_Paciente",
    how="inner"
)
display(df_consolidado.head())
```

2. **Criação de Colunas Temporais**: Extraia o Mês (número ou nome) e o Dia da Semana (onde 0=Segunda, 6=Domingo, ou usando .dt.day_name()).


```python
df_atendimentos["Mes"] = df_atendimentos["Data_Hora"].dt.strftime("%B")
display(df_atendimentos["Mes"].head()) 
```

```python
df_atendimentos["Semana"] = df_atendimentos["Data_Hora"].dt.strftime("%A")
display(df_atendimentos["Semana"].head())
```
